```r
data <- read.csv("data1.csv")
```

```r
print(data)
```

Logit Models:

```r
m1_logit <- glm(Pass/Total ~ Treat1 + Dose1 + Treat2 + Dose2, weights = Total, family = binomial(link = "logit"), data= data)
```

```r
print(summary(m1_logit))
```

```r
m2_logit <- glm(Pass/Total ~ Treat1:Dose1 + Treat2:Dose2, weights = Total, family = binomial(link = "logit"), data= data)
```

```r
print(summary(m2_logit))
```

```r
m3_logit <- glm(Pass/Total ~ Treat1*Dose1 + Treat2*Dose2, weights = Total, family = binomial(link = "logit"), data= data)
```

```r
print(summary(m3_logit))
```

```r
m4_logit <- glm(Pass/Total ~ Treat1:Dose1 + Treat2:Dose2 + Treat1 + Treat2, weights = Total, family = binomial(link = "logit"), data= data)
```

```r
print(summary(m4_logit))
```

Probit Models:

```r
m1_probit <- glm(Pass/Total ~ Treat1 + Dose1 + Treat2 + Dose2, weights = Total, family = binomial(link = "probit"), data= data)
print(summary(m1_probit))
```

```r
m2_probit <- glm(Pass/Total ~ Treat1:Dose1 + Treat2:Dose2, weights = Total, family = binomial(link = "probit"), data= data)
print(summary(m2_probit))
```

```r
m3_probit <- glm(Pass/Total ~ Treat1*Dose1 + Treat2*Dose2, weights = Total, family = binomial(link = "probit"), data= data)
print(summary(m3_probit))
```

```r
m4_probit <- glm(Pass/Total ~ Treat1:Dose1 + Treat2:Dose2 + Treat1 + Treat2, weights = Total, family = binomial(link = "probit"), data= data)
print(summary(m4_probit))
```

LogLog Models:

```r
m1_loglog <- glm(Pass/Total ~ Treat1 + Dose1 + Treat2 + Dose2, weights = Total, family = binomial(link = "cloglog"), data= data)
print(summary(m1_loglog))
```

```r
m2_loglog <- glm(Pass/Total ~ Treat1:Dose1 + Treat2:Dose2, weights = Total, family = binomial(link = "cloglog"), data= data)
print(summary(m2_loglog))
```

```r
m3_loglog <- glm(Pass/Total ~ Treat1*Dose1 + Treat2*Dose2, weights = Total, family = binomial(link = "cloglog"), data= data)
print(summary(m3_loglog))
```

```r
m4_loglog <- glm(Pass/Total ~ Treat1:Dose1 + Treat2:Dose2 + Treat1 + Treat2, weights = Total, family = binomial(link = "cloglog"), data= data)
print(summary(m4_loglog))
```
